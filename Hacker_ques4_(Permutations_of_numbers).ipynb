```python
a,b,c,n = input('write value of x\n'),input('write value of y\n'),input('write value of z\n'),input('write value of n\n')
list_of_lists = []

for num1 in range(int(a) + 1):
  for num2 in range(int(b) + 1):
    for num3 in range(int(c) + 1):
      if num1 + num2 + num3 != int(n):
        list_of_lists.append([num1,num2,num3])

print(list_of_lists)
```

```python
# We can do the same using List comprehension as Shown below:- 
```

```python
a,b,c,n = 1,1,1,2
print([[x,y,z] for x in range(a + 1) for y in range(b + 1) for z in range(c + 1) if x + y + z != n])
```
