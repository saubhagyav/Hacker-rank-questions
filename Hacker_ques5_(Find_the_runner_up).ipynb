```python
  n = int(input('write the value of n?\n'))
  list_a = []
  for i in range(n):
    list_a.append(input('write the series of scores\n'))
  
    
  list_a.sort()
  y = list_a.pop(-1)
  iteration = 1
  for elem in list_a:
    if y == list_a[-1]:
      list_a.pop()
      iteration += 1
      if iteration == n-1:
        print(f"The runner-up player's score is {list_a[0]}")
      continue
    else:
      print(f"The runner-up player's score is {list_a[-1]}")
      break
```
